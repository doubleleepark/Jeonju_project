```python
import re
def clean_text(text):
    cleaned_text = re.sub('[a-zA-Z]' , '', text)
    cleaned_text = re.sub('[\{\}\[\]\/?.,;:|\)*~`!^\-_+<>@\#$%&\\\=\(\'\"]',
                          '', cleaned_text)
 
    return cleaned_text
```

```python
text = clean_text(text)
```

```python
wordList = text.split()
worduniq = list(set(wordList))
wordCount = {}
for w in worduniq:
    wordCount[w] = wordList.count(w)
```
